```python
import pandas as pd
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
independent
```

```python
dependent = dataset[['Profit']]
dependent
```

```python
#gradiantboost Regressor
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
param_grid = {'criterion':['squared_error','friedman_mse'],'loss':['absolute_error','squared_error','quantile','huber'],
               'max_features':['none','sqrt','log2'],'n_estimators':[30,40,50,100]}
grid = GridSearchCV(GradientBoostingRegressor(),param_grid,verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
print(grid.best_params_)
re=grid.cv_results_
print(re)
```

```python
print("The best R_score value Parameters:{}".format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(re)
table
```

```python
rd_spend = float(input("Enter R&D spend value:"))
admin_value = float(input("Enter Administration spend:"))
marketing_spend = float(input("Enter Marketing spend:"))
state_input1 = int(input("Enter 0 or 1 for State_florida:"))
state_input2 = int(input("Enter 0 or 1 for State_newYork:"))
Future_prediction=grid.predict([[rd_spend,admin_value,marketing_spend,state_input1,state_input2]])
print("Future_Prediction={}".format(Future_Prediction))
```

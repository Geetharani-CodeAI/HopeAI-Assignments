```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
independent = dataset[['age','bmi','children','sex_male','smoker_yes']]
independent
```

```python
dependent= dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid = {'criterion':['squared_error','absolute_error','friedman_mse','poisson'],
              'max_features':['none','sqrt','log2'],'n_estimators':[10,100]}
grid = GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
#Printing the best Parameter after tuning
print(grid.best_params_)
re = grid.cv_results_
print(re)
```

```python
# Print Classification report

print("The best R_score value parameters are {}".format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(re)
table
```

```python
age = float(input("Age:"))
bmi = float(input("BMI:"))
children = float(input("Children:"))
sex_male = int(input("sex_male 0 or 1:"))
smoker_yes = int(input("smoker_yes 0 or 1"))
Future_Prediction = grid.predict([[age,bmi,children,sex_male,smoker_yes]])
print("Future_Prediction={}".format(Future_Prediction))
```
